```python
import pandas as pd
from sklearn.impute import SimpleImputer
```

```python
data = pd.read_csv("books.csv")
data.head()
```

```python
#sum of null values
missing_values = data.isnull().values.sum()
print("Total number of missing values: ",missing_values)

#percentage of null values
missing_percentage = (data.isnull().sum() / len(data)) * 100
print("Percentage of missing values for training data:")
print(missing_percentage)

#highest column percentage of null values
highest_missing_percentage = missing_percentage.idxmax()
print(f"The column with the highest missing values percentage is: {highest_missing_percentage}")
```

```python
numImpute = SimpleImputer(strategy='mean')
catImpute = SimpleImputer(strategy='most_frequent')
numerical_columns = data.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = [col for col in data.columns if pd.api.types.is_categorical_dtype(data[col]) or data[col].dtype == 'object']
data[numerical_columns] = numImpute.fit_transform(data[numerical_columns])
data[categorical_cols] = catImpute.fit_transform(data[categorical_cols])

#sum of null values
missing_values = data.isnull().values.sum()
print("Total number of missing values: ",missing_values)
```

```python
harry_potter_books = data[data['original_title'].str.contains('Harry Potter', case=False) | 
                          data['title'].str.contains('Harry Potter', case=False)]
harry_potter_books.describe()
print(harry_potter_books)
```

```python
most_selling_harry_potter_books = harry_potter_books.sort_values(by='work_ratings_count', ascending=False).head(1)['original_title']
print("The most selling harry potter book:",most_selling_harry_potter_books)
```

```python
average_rating_harry_potter = harry_potter_books['average_rating'].mean()
print("Average rating harry potter:",average_rating_harry_potter)
```
